### Menggabungkan kolom nama negara dan tingkat pengangguran

```python
df_unemployment = pd.concat([df['Countries and areas'], df['Unemployment_Rate']], axis=1)
df_unemployment.info()
```

### Kategorisasi kolom tingkat pengangguran berdasarkan nilai kuartilnya

```python
kuartil1 = df_unemployment['Unemployment_Rate'].quantile(0.25)
kuartil2 = df_unemployment['Unemployment_Rate'].quantile(0.50)
kuartil3 = df_unemployment['Unemployment_Rate'].quantile(0.75)

def kategorisasi_kuartil(nilai):
    if nilai <= kuartil1:
        return 'D'
    elif nilai <= kuartil2:
        return 'C'
    elif nilai <= kuartil3:
        return 'B'
    else:
        return 'A'

df_unemployment['Unemployment_Rate'] = df_unemployment['Unemployment_Rate'].apply(kategorisasi_kuartil)
df_unemployment
```

### Memeriksa outlier setiap kolom menggunakan boxplot dan menyimpan index yang terdapat ouliernya 

```python
numeric_cols = df.select_dtypes(include=['number'])

num_cols = 2
num_rows = -(-len(numeric_cols.columns) // num_cols)

outlier_indices = {}

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))
axes = axes.flatten()

for i, col in enumerate(numeric_cols.columns):
    axes[i].boxplot(numeric_cols[col], vert=False)
    axes[i].set_title(col)

    # Hitung outlier
    Q1 = numeric_cols[col].quantile(0.25)
    Q3 = numeric_cols[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outlier_indices[col] = numeric_cols[col][(numeric_cols[col] < lower_bound) | (numeric_cols[col] > upper_bound)].index.tolist()

for i in range(len(numeric_cols.columns), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()

print("Index Outlier:")
for col, indices in outlier_indices.items():
    print(f"'{col}': {indices}")
    
outlier_data = {col: outlier_indices[col] for col in outlier_indices}
```

### Mengatasi outlier dengan transformasi data menggunakan winsorized

```python
for col, indices in outlier_data.items():
    df[col] = stats.mstats.winsorize(df[col], limits=[0.1, 0.1]) 
df
```
